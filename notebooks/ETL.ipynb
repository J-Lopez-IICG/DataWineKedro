```python
# Cargar los datos usando el 'catalog' que creamos en la celda anterior
data = catalog.load("processed_wine_data") # type: ignore

# Imprimir los primeros 5 registros
data.head()
```

```python
# 1. Revisar si hay valores nulos en cada columna
print("Valores nulos por columna antes de la limpieza:")
print(data.isnull().sum())

# 2. Eliminar filas que contengan al menos un valor nulo
# .dropna() es el comando para eliminar filas (o columnas) con valores NaN
# 'inplace=True' modifica el DataFrame 'data' directamente
data.dropna(inplace=True)

print("\nValores nulos por columna después de la limpieza:")
print(data.isnull().sum())

# 3. Revisar si hay filas duplicadas
print(f"\nNúmero de filas duplicadas encontradas: {data.duplicated().sum()}")

# 4. Eliminar filas duplicadas
# 'inplace=True' modifica el DataFrame 'data' directamente
data.drop_duplicates(inplace=True)

print(f"Número de filas duplicadas después de la limpieza: {data.duplicated().sum()}")

# 5. Mostrar información general del DataFrame limpio
print("\nInformación del DataFrame después de la limpieza:")
data.info()

# 6. Verificar que todas las columnas sean numéricas (int o float)
print("\n--- Verificación de Tipos de Datos ---")
import pandas as pd
from pandas.api.types import is_numeric_dtype

non_numeric_columns = []
for column in data.columns:
    # Comprobamos si el tipo de dato de la columna NO es numérico
    if not is_numeric_dtype(data[column]):
        non_numeric_columns.append(column)

if not non_numeric_columns:
    print("¡Excelente! Todas las columnas tienen un formato numérico (int o float).")
else:
    print("¡Atención! Se encontraron columnas con formato no numérico:")
    for col in non_numeric_columns:
        print(f"- Columna: '{col}', Tipo: {data[col].dtype}")
```

```python
# Crear un diccionario para mapear los nombres de las columnas de inglés a español
# Es una buena práctica usar minúsculas y guiones bajos para los nombres de las columnas
column_mapping = {
    'fixed acidity': 'Acidez_Fija',
    'volatile acidity': 'Acidez_Volatil',
    'citric acid': 'Acido_Citrico',
    'residual sugar': 'Azucar_Residual',
    'chlorides': 'Cloruros',
    'free sulfur dioxide': 'Dioxido_Azufre_Libre',
    'total sulfur dioxide': 'Dioxido_Azufre_Total',
    'density': 'Densidad',
    'pH': 'ph',
    'sulphates': 'Sulfatos',
    'alcohol': 'Alcohol',
    'quality': 'Calidad'  # No olvidar la variable de salida
}

# Renombrar las columnas del DataFrame
data.rename(columns=column_mapping, inplace=True)

# Verificar que los nombres de las columnas se han actualizado correctamente
print("Nombres de las columnas actualizados:")
print(data.columns)

# Opcional: Mostrar las primeras filas con los nuevos nombres
print("\nDataFrame con las nuevas columnas:")
data.head()
```

```python
# Crear la nueva columna 'Calidad_Binaria'
# Usaremos una condición booleana. Esto crea una serie de True/False.
# Al convertirla a entero con .astype(int), True se convierte en 1 y False en 0.
data['Calidad_Binaria'] = (data['Calidad'] >= 7).astype(int)

# Verificar la distribución de la nueva columna (opcional pero recomendado)
print("Distribución de la nueva columna 'Calidad_Binaria':")
print(data['Calidad_Binaria'].value_counts())

columnas_a_mantener = [
    'Acidez_Fija',
    'Acidez_Volatil',
    'Acido_Citrico',
    'Azucar_Residual',
    'Cloruros',
    'Dioxido_Azufre_Libre',
    'Dioxido_Azufre_Total',
    'Densidad',
    'ph',
    'Sulfatos',
    'Alcohol',
    #'Calidad',
    'Calidad_Binaria',
]
data_seleccionada = data[columnas_a_mantener].copy()

# Mostrar las primeras filas del nuevo DataFrame con las columnas seleccionadas
print("DataFrame con las columnas seleccionadas:")
print(data_seleccionada.head())

# Mostrar la información del nuevo DataFrame para confirmar las columnas y tipos de datos
print("\nInformación del DataFrame con columnas seleccionadas:")
data_seleccionada.info()
```

```python
# Importar las librerías necesarias para el modelado
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report, roc_auc_score, roc_curve
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd # Asegurarse de que pandas esté importado

print("--- Preparación de Datos para el Modelo ---")

# 1. Separar las características (X) de la variable objetivo (y)
# La variable objetivo es 'Calidad_Binaria'
X = data_seleccionada.drop('Calidad_Binaria', axis=1) # Todas las columnas excepto 'Calidad_Binaria'
y = data_seleccionada['Calidad_Binaria'] # Solo la columna 'Calidad_Binaria'

print(f"Dimensiones de X (características): {X.shape}")
print(f"Dimensiones de y (variable objetivo): {y.shape}")

# 2. Dividir los datos en conjuntos de entrenamiento y prueba
# test_size=0.30 significa que el 30% de los datos se usarán para prueba
# random_state=42 asegura que la división sea la misma cada vez que ejecutes el código
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42, stratify=y)

print(f"Dimensiones de X_train: {X_train.shape}")
print(f"Dimensiones de X_test: {X_test.shape}")
print(f"Dimensiones de y_train: {y_train.shape}")
print(f"Dimensiones de y_test: {y_test.shape}")

print("\n--- Entrenamiento del Modelo de Árbol de Decisión ---")

# 3. Inicializar el modelo de Árbol de Decisión
# random_state=42 para reproducibilidad
# Puedes ajustar otros parámetros como max_depth, min_samples_leaf, etc.
model_dt = DecisionTreeClassifier(random_state=42)

# 4. Entrenar el modelo con los datos de entrenamiento
model_dt.fit(X_train, y_train)

print("Modelo de Árbol de Decisión entrenado exitosamente.")

print("\n--- Evaluación del Modelo ---")

# 5. Realizar predicciones sobre el conjunto de prueba
y_pred = model_dt.predict(X_test)
y_pred_proba = model_dt.predict_proba(X_test)[:, 1] # Probabilidades para la clase positiva (1)

# 6. Calcular y mostrar métricas de evaluación
print(f"Precisión (Accuracy): {accuracy_score(y_test, y_pred):.4f}")
print("\nReporte de Clasificación:")
print(classification_report(y_test, y_pred))

# Calcular y mostrar el AUC (Area Under the Curve)
auc_score = roc_auc_score(y_test, y_pred_proba)
print(f"AUC (Area Under the Curve): {auc_score:.4f}")

# 7. Visualizar la Curva ROC
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'Curva ROC (AUC = {auc_score:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', label='Clasificador Aleatorio')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Tasa de Falsos Positivos (FPR)')
plt.ylabel('Tasa de Verdaderos Positivos (TPR)')
plt.title('Curva ROC - Árbol de Decisión')
plt.legend(loc="lower right")
plt.grid(True)
plt.show()

print("\n--- Interpretación del Árbol de Decisión ---")
from sklearn.tree import plot_tree
plt.figure(figsize=(20,10))
plot_tree(model_dt, feature_names=X.columns, class_names=['Malo', 'Bueno'], filled=True, rounded=True)
plt.title("Árbol de Decisión")
plt.show()

# También puedes ver la importancia de las características
importancia_caracteristicas = pd.DataFrame({
    'Caracteristica': X.columns,
    'Importancia': model_dt.feature_importances_
}).sort_values(by='Importancia', ascending=False)

print("\nImportancia de las Características:")
print(importancia_caracteristicas)

plt.figure(figsize=(10, 6))
sns.barplot(x='Importancia', y='Caracteristica', data=importancia_caracteristicas)
plt.title('Importancia de las Características en el Árbol de Decisión')
plt.show()
```
