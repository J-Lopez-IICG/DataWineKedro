```python
# Cargar los datos usando el 'catalog' que creamos en la celda anterior
data = catalog.load("processed_wine_data") # type: ignore

# Imprimir los primeros 5 registros
data.head()
```

## 2. Limpieza y Verificaci√≥n de Datos

Antes de cualquier an√°lisis o modelado, es fundamental asegurar la calidad de los datos. En esta secci√≥n, realizaremos los siguientes pasos:
1.  **Verificar Valores Nulos:** Buscaremos y eliminaremos cualquier fila que contenga datos faltantes.
2.  **Verificar Duplicados:** Identificaremos y eliminaremos registros duplicados para evitar sesgos en el modelo.
3.  **Verificar Tipos de Datos:** Nos aseguraremos de que todas las columnas de caracter√≠sticas sean num√©ricas, lo cual es un requisito para los modelos que utilizaremos.

```python
# 1. Revisar si hay valores nulos en cada columna
print("Valores nulos por columna antes de la limpieza:")
print(data.isnull().sum())

# 2. Eliminar filas que contengan al menos un valor nulo
# .dropna() es el comando para eliminar filas (o columnas) con valores NaN
# 'inplace=True' modifica el DataFrame 'data' directamente
data.dropna(inplace=True)

print("\nValores nulos por columna despu√©s de la limpieza:")
print(data.isnull().sum())

# 3. Revisar si hay filas duplicadas
print(f"\nN√∫mero de filas duplicadas encontradas: {data.duplicated().sum()}")

# 4. Eliminar filas duplicadas
# 'inplace=True' modifica el DataFrame 'data' directamente
data.drop_duplicates(inplace=True)

print(f"N√∫mero de filas duplicadas despu√©s de la limpieza: {data.duplicated().sum()}")

# 5. Mostrar informaci√≥n general del DataFrame limpio
print("\nInformaci√≥n del DataFrame despu√©s de la limpieza:")
data.info()

# 6. Verificar que todas las columnas sean num√©ricas (int o float)
print("\n--- Verificaci√≥n de Tipos de Datos ---")
import pandas as pd
from pandas.api.types import is_numeric_dtype

non_numeric_columns = []
for column in data.columns:
    # Comprobamos si el tipo de dato de la columna NO es num√©rico
    if not is_numeric_dtype(data[column]):
        non_numeric_columns.append(column)

if not non_numeric_columns:
    print("¬°Excelente! Todas las columnas tienen un formato num√©rico (int o float).")
else:
    print("¬°Atenci√≥n! Se encontraron columnas con formato no num√©rico:")
    for col in non_numeric_columns:
        print(f"- Columna: '{col}', Tipo: {data[col].dtype}")
```

## 3. Renombrar Columnas

Para mejorar la legibilidad y la coherencia del c√≥digo, traduciremos los nombres de las columnas del ingl√©s al espa√±ol, utilizando un formato est√°ndar de `snake_case` (por ejemplo, `fixed acidity` se convertir√° en `acidez_fija`).

```python
# Crear un diccionario para mapear los nombres de las columnas de ingl√©s a espa√±ol
# Es una buena pr√°ctica usar min√∫sculas y guiones bajos para los nombres de las columnas
column_mapping = {
    'fixed acidity': 'Acidez_Fija',
    'volatile acidity': 'Acidez_Volatil',
    'citric acid': 'Acido_Citrico',
    'residual sugar': 'Azucar_Residual',
    'chlorides': 'Cloruros',
    'free sulfur dioxide': 'Dioxido_Azufre_Libre',
    'total sulfur dioxide': 'Dioxido_Azufre_Total',
    'density': 'Densidad',
    'pH': 'ph',
    'sulphates': 'Sulfatos',
    'alcohol': 'Alcohol',
    'quality': 'Calidad'  # No olvidar la variable de salida
}

# Renombrar las columnas del DataFrame
data.rename(columns=column_mapping, inplace=True)

# Verificar que los nombres de las columnas se han actualizado correctamente
print("Nombres de las columnas actualizados:")
print(data.columns)

# Opcional: Mostrar las primeras filas con los nuevos nombres
print("\nDataFrame con las nuevas columnas:")
data.head()
```

## 4. Ingenier√≠a de Caracter√≠sticas: Creaci√≥n de la Variable Objetivo

La columna `Calidad` original es una variable categ√≥rica ordinal con valores del 3 al 8. Para simplificar el problema y convertirlo en una tarea de clasificaci√≥n binaria, crearemos una nueva columna llamada `Calidad_Binaria`:
-   **1 (Buena Calidad):** Si la calidad original es 7 o superior.
-   **0 (Mala Calidad):** Si la calidad original es inferior a 7.

Este enfoque es com√∫n cuando el objetivo es distinguir entre vinos "normales" y vinos "premium" o de alta calidad.

```python
# Crear la nueva columna 'Calidad_Binaria'
# Usaremos una condici√≥n booleana. Esto crea una serie de True/False.
# Al convertirla a entero con .astype(int), True se convierte en 1 y False en 0.
data['Calidad_Binaria'] = (data['Calidad'] >= 7).astype(int)

# Verificar la distribuci√≥n de la nueva columna (opcional pero recomendado)
print("Distribuci√≥n de la nueva columna 'Calidad_Binaria':")
print(data['Calidad_Binaria'].value_counts())

columnas_a_mantener = [
    'Acidez_Fija',
    'Acidez_Volatil',
    'Acido_Citrico',
    'Azucar_Residual',
    'Cloruros',
    'Dioxido_Azufre_Libre',
    'Dioxido_Azufre_Total',
    'Densidad',
    'ph',
    'Sulfatos',
    'Alcohol',
    #'Calidad',
    'Calidad_Binaria',
]
data_seleccionada = data[columnas_a_mantener].copy()

# Mostrar las primeras filas del nuevo DataFrame con las columnas seleccionadas
print("DataFrame con las columnas seleccionadas:")
print(data_seleccionada.head())

# Mostrar la informaci√≥n del nuevo DataFrame para confirmar las columnas y tipos de datos
print("\nInformaci√≥n del DataFrame con columnas seleccionadas:")
data_seleccionada.info()
```

## 5. Preparaci√≥n de Datos para el Modelado

En este paso, dividiremos nuestros datos para el entrenamiento y la evaluaci√≥n del modelo:
1.  **Separaci√≥n de Caracter√≠sticas (X) y Objetivo (y):** Aislaremos la columna `Calidad_Binaria` (nuestra variable objetivo `y`) del resto de las caracter√≠sticas (`X`).
2.  **Divisi√≥n en Entrenamiento y Prueba:** Dividiremos el conjunto de datos en un 70% para entrenamiento (`X_train`, `y_train`) y un 30% para prueba (`X_test`, `y_test`). El conjunto de prueba no se utilizar√° durante el entrenamiento y servir√° para evaluar el rendimiento final del modelo en datos no vistos. Usamos `stratify=y` para mantener la misma proporci√≥n de clases en ambos conjuntos, lo cual es crucial en datasets desbalanceados.

```python
# Importar las librer√≠as necesarias para el modelado
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report, roc_auc_score, roc_curve
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd # Asegurarse de que pandas est√© importado

print("--- Preparaci√≥n de Datos para el Modelo ---")

# 1. Separar las caracter√≠sticas (X) de la variable objetivo (y)
# La variable objetivo es 'Calidad_Binaria'
X = data_seleccionada.drop('Calidad_Binaria', axis=1) # Todas las columnas excepto 'Calidad_Binaria'
y = data_seleccionada['Calidad_Binaria'] # Solo la columna 'Calidad_Binaria'

print(f"Dimensiones de X (caracter√≠sticas): {X.shape}")
print(f"Dimensiones de y (variable objetivo): {y.shape}")

# 2. Dividir los datos en conjuntos de entrenamiento y prueba
# test_size=0.30 significa que el 30% de los datos se usar√°n para prueba
# random_state=42 asegura que la divisi√≥n sea la misma cada vez que ejecutes el c√≥digo
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42, stratify=y)

print(f"Dimensiones de X_train: {X_train.shape}")
print(f"Dimensiones de X_test: {X_test.shape}")
print(f"Dimensiones de y_train: {y_train.shape}")
print(f"Dimensiones de y_test: {y_test.shape}")
```

## 6. Modelo 1: √Årbol de Decisi√≥n (L√≠nea Base)

Comenzamos con un **√Årbol de Decisi√≥n** como nuestro modelo de l√≠nea base. Es un modelo simple, r√°pido de entrenar y f√°cil de interpretar visualmente. Nos dar√° una primera idea del rendimiento que podemos esperar y de las caracter√≠sticas m√°s importantes seg√∫n este algoritmo.

```python
print("\n--- Entrenamiento del Modelo de √Årbol de Decisi√≥n ---")

# 3. Inicializar el modelo de √Årbol de Decisi√≥n
# random_state=42 para reproducibilidad
# Puedes ajustar otros par√°metros como max_depth, min_samples_leaf, etc.
model_dt = DecisionTreeClassifier(random_state=42)

# 4. Entrenar el modelo con los datos de entrenamiento
model_dt.fit(X_train, y_train)

print("Modelo de √Årbol de Decisi√≥n entrenado exitosamente.")

print("\n--- Evaluaci√≥n del Modelo ---")

# 5. Realizar predicciones sobre el conjunto de prueba
y_pred = model_dt.predict(X_test)
y_pred_proba = model_dt.predict_proba(X_test)[:, 1] # type: ignore

# 6. Calcular y mostrar m√©tricas de evaluaci√≥n
print(f"Precisi√≥n (Accuracy): {accuracy_score(y_test, y_pred):.4f}")
print("\nReporte de Clasificaci√≥n:")
print(classification_report(y_test, y_pred))

# Calcular y mostrar el AUC (Area Under the Curve)
auc_score = roc_auc_score(y_test, y_pred_proba)
print(f"AUC (Area Under the Curve): {auc_score:.4f}")

# 7. Visualizar la Curva ROC
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'Curva ROC (AUC = {auc_score:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', label='Clasificador Aleatorio')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Tasa de Falsos Positivos (FPR)')
plt.ylabel('Tasa de Verdaderos Positivos (TPR)')
plt.title('Curva ROC - √Årbol de Decisi√≥n')
plt.legend(loc="lower right")
plt.grid(True)
plt.show()

print("\n--- Interpretaci√≥n del √Årbol de Decisi√≥n ---")
from sklearn.tree import plot_tree
plt.figure(figsize=(20,10))
plot_tree(model_dt, feature_names=X.columns, class_names=['Malo', 'Bueno'], filled=True, rounded=True)
plt.title("√Årbol de Decisi√≥n")
plt.show()

# Tambi√©n puedes ver la importancia de las caracter√≠sticas
importancia_caracteristicas = pd.DataFrame({
    'Caracteristica': X.columns,
    'Importancia': model_dt.feature_importances_
}).sort_values(by='Importancia', ascending=False)

print("\nImportancia de las Caracter√≠sticas:")
print(importancia_caracteristicas)

plt.figure(figsize=(10, 6))
sns.barplot(x='Importancia', y='Caracteristica', data=importancia_caracteristicas)
plt.title('Importancia de las Caracter√≠sticas en el √Årbol de Decisi√≥n')
plt.show()
```

## 7. Optimizaci√≥n del √Årbol de Decisi√≥n con GridSearchCV

El rendimiento de un modelo depende en gran medida de sus **hiperpar√°metros**. Para mejorar nuestro √Årbol de Decisi√≥n, utilizaremos `GridSearchCV`. Esta t√©cnica prueba sistem√°ticamente una "cuadr√≠cula" de diferentes combinaciones de hiperpar√°metros (como `max_depth` o `min_samples_leaf`) y, mediante validaci√≥n cruzada, encuentra la combinaci√≥n que maximiza el rendimiento, en nuestro caso, el `roc_auc`.

```python
# Importar GridSearchCV para el ajuste de hiperpar√°metros
from sklearn.model_selection import GridSearchCV

print("--- Ajuste de Hiperpar√°metros para el √Årbol de Decisi√≥n (GridSearchCV) ---")

# 1. Definir los hiperpar√°metros que quieres probar y sus rangos
# Estos son algunos de los par√°metros m√°s comunes para un √Årbol de Decisi√≥n
param_grid = {
    'max_depth': [3, 5, 7, 10, None], # Profundidad m√°xima del √°rbol. None significa sin l√≠mite.
    'min_samples_leaf': [1, 5, 10, 20], # N√∫mero m√≠nimo de muestras requeridas para ser un nodo hoja.
    'min_samples_split': [2, 5, 10], # N√∫mero m√≠nimo de muestras requeridas para dividir un nodo.
    'criterion': ['gini', 'entropy'] # Funci√≥n para medir la calidad de una divisi√≥n.
}

# 2. Inicializar el modelo de √Årbol de Decisi√≥n base
# Usamos el mismo random_state para reproducibilidad
model_dt_base = DecisionTreeClassifier(random_state=42)

# 3. Configurar GridSearchCV
# estimator: el modelo que queremos ajustar
# param_grid: el diccionario de par√°metros a probar
# cv: n√∫mero de pliegues para la validaci√≥n cruzada (5 es un valor com√∫n)
# scoring: la m√©trica que queremos optimizar (roc_auc es ideal para clasificaci√≥n binaria desbalanceada)
# n_jobs: -1 usa todos los n√∫cleos de la CPU para acelerar el proceso
# verbose: 1 muestra el progreso
grid_search = GridSearchCV(estimator=model_dt_base, param_grid=param_grid,
                           cv=5, scoring='roc_auc', n_jobs=-1, verbose=1)

# 4. Entrenar GridSearchCV para encontrar la mejor combinaci√≥n de par√°metros
# Esto puede tardar un poco dependiendo del tama√±o de tu param_grid y tus datos
grid_search.fit(X_train, y_train)

# 5. Mostrar los mejores par√°metros encontrados y la mejor puntuaci√≥n de AUC
print(f"\nMejores par√°metros encontrados: {grid_search.best_params_}")
print(f"Mejor AUC en validaci√≥n cruzada: {grid_search.best_score_:.4f}")

# 6. Obtener el mejor modelo (el √Årbol de Decisi√≥n con los mejores par√°metros)
best_model_dt = grid_search.best_estimator_

print("\n--- Evaluaci√≥n del Mejor Modelo de √Årbol de Decisi√≥n ---")

# 7. Realizar predicciones con el mejor modelo sobre el conjunto de prueba
y_pred_best = best_model_dt.predict(X_test)
y_pred_proba_best = best_model_dt.predict_proba(X_test)[:, 1] # type: ignore # Probabilidades para la clase positiva (1)

# 8. Calcular y mostrar m√©tricas de evaluaci√≥n para el mejor modelo
print(f"Precisi√≥n (Accuracy) del mejor modelo: {accuracy_score(y_test, y_pred_best):.4f}")
print("\nReporte de Clasificaci√≥n del mejor modelo:")
print(classification_report(y_test, y_pred_best))

auc_score_best = roc_auc_score(y_test, y_pred_proba_best)
print(f"AUC (Area Under the Curve) del mejor modelo: {auc_score_best:.4f}")

# 9. Visualizar la Curva ROC del mejor modelo
fpr_best, tpr_best, thresholds_best = roc_curve(y_test, y_pred_proba_best)

plt.figure(figsize=(8, 6))
plt.plot(fpr_best, tpr_best, color='darkorange', lw=2, label=f'Curva ROC (AUC = {auc_score_best:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', label='Clasificador Aleatorio')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Tasa de Falsos Positivos (FPR)')
plt.ylabel('Tasa de Verdaderos Positivos (TPR)')
plt.title('Curva ROC - Mejor √Årbol de Decisi√≥n (GridSearchCV)')
plt.legend(loc="lower right")
plt.grid(True)
plt.show()

# Opcional: Visualizar la importancia de las caracter√≠sticas del mejor modelo
importancia_caracteristicas_best = pd.DataFrame({
    'Caracteristica': X.columns,
    'Importancia': best_model_dt.feature_importances_
}).sort_values(by='Importancia', ascending=False)

print("\nImportancia de las Caracter√≠sticas del Mejor Modelo:")
print(importancia_caracteristicas_best)

plt.figure(figsize=(10, 6))
sns.barplot(x='Importancia', y='Caracteristica', data=importancia_caracteristicas_best)
plt.title('Importancia de las Caracter√≠sticas en el Mejor √Årbol de Decisi√≥n')
plt.show()
```

## 8. Modelo 2: Random Forest

El siguiente modelo es el **Random Forest**, un m√©todo de *ensemble* que construye m√∫ltiples √°rboles de decisi√≥n durante el entrenamiento y combina sus resultados. Generalmente, esto conduce a un modelo m√°s robusto, con mayor precisi√≥n y menor riesgo de sobreajuste que un √∫nico √°rbol de decisi√≥n.

Para manejar el desbalance de clases, utilizaremos el par√°metro `class_weight='balanced'`, que ajusta los pesos de las muestras de forma inversamente proporcional a la frecuencia de sus clases.

```python
# Importar el clasificador Random Forest
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, roc_auc_score, roc_curve
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd # Asegurarse de que pandas est√© importado

print("--- Entrenamiento del Modelo Random Forest ---")

# 1. Inicializar el modelo Random Forest
# n_estimators: n√∫mero de √°rboles en el bosque (un valor com√∫n es 100 o m√°s)
# random_state: para reproducibilidad
# class_weight='balanced': importante para manejar el desbalance de clases,
#                          ajusta autom√°ticamente los pesos de las clases.
model_rf = RandomForestClassifier(n_estimators=100, random_state=42, class_weight='balanced')

# 2. Entrenar el modelo con los datos de entrenamiento
model_rf.fit(X_train, y_train)

print("Modelo Random Forest entrenado exitosamente.")

print("\n--- Evaluaci√≥n del Modelo Random Forest ---")

# 3. Realizar predicciones sobre el conjunto de prueba
y_pred_rf = model_rf.predict(X_test)
y_pred_proba_rf = model_rf.predict_proba(X_test)[:, 1] # Probabilidades para la clase positiva (1)

# 4. Calcular y mostrar m√©tricas de evaluaci√≥n
print(f"Precisi√≥n (Accuracy) Random Forest: {accuracy_score(y_test, y_pred_rf):.4f}")
print("\nReporte de Clasificaci√≥n Random Forest:")
print(classification_report(y_test, y_pred_rf))

auc_score_rf = roc_auc_score(y_test, y_pred_proba_rf)
print(f"AUC (Area Under the Curve) Random Forest: {auc_score_rf:.4f}")

# 5. Visualizar la Curva ROC del modelo Random Forest
fpr_rf, tpr_rf, thresholds_rf = roc_curve(y_test, y_pred_proba_rf)

plt.figure(figsize=(8, 6))
plt.plot(fpr_rf, tpr_rf, color='darkgreen', lw=2, label=f'Curva ROC Random Forest (AUC = {auc_score_rf:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', label='Clasificador Aleatorio')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Tasa de Falsos Positivos (FPR)')
plt.ylabel('Tasa de Verdaderos Positivos (TPR)')
plt.title('Curva ROC - Random Forest')
plt.legend(loc="lower right")
plt.grid(True)
plt.show()

print("\n--- Importancia de las Caracter√≠sticas (Random Forest) ---")

# 6. Ver la importancia de las caracter√≠sticas
importancia_caracteristicas_rf = pd.DataFrame({
    'Caracteristica': X.columns,
    'Importancia': model_rf.feature_importances_
}).sort_values(by='Importancia', ascending=False)

print(importancia_caracteristicas_rf)

plt.figure(figsize=(10, 6))
sns.barplot(x='Importancia', y='Caracteristica', data=importancia_caracteristicas_rf)
plt.title('Importancia de las Caracter√≠sticas en Random Forest')
plt.show()
```

## 9. Optimizaci√≥n de Random Forest con GridSearchCV

Al igual que con el √Årbol de Decisi√≥n, ahora optimizaremos los hiperpar√°metros del Random Forest. La cuadr√≠cula de b√∫squeda incluir√° par√°metros clave como el n√∫mero de √°rboles (`n_estimators`), la profundidad m√°xima (`max_depth`) y los criterios de divisi√≥n. El objetivo es encontrar la configuraci√≥n que nos d√© el mejor rendimiento posible con este algoritmo.

```python
# Importar GridSearchCV para el ajuste de hiperpar√°metros
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier # Asegurarse de que est√© importado

print("--- Ajuste de Hiperpar√°metros para Random Forest (GridSearchCV) ---")

# 1. Definir los hiperpar√°metros que quieres probar y sus rangos para Random Forest
# Es importante no hacer la cuadr√≠cula demasiado grande, ya que Random Forest es m√°s lento que un solo √°rbol.
param_grid_rf = {
    'n_estimators': [50, 100, 200], # N√∫mero de √°rboles
    'max_depth': [5, 10, 15, None], # Profundidad m√°xima de cada √°rbol
    'min_samples_leaf': [1, 5, 10], # M√≠nimo de muestras en un nodo hoja
    'min_samples_split': [2, 5, 10], # M√≠nimo de muestras para dividir un nodo
    'criterion': ['gini', 'entropy'], # Criterio de divisi√≥n
    'class_weight': ['balanced'] # Mantener el manejo del desbalance de clases
}

# 2. Inicializar el modelo Random Forest base
# Usamos el mismo random_state para reproducibilidad
model_rf_base = RandomForestClassifier(random_state=42)

# 3. Configurar GridSearchCV para Random Forest
# cv=5: validaci√≥n cruzada de 5 pliegues
# scoring='roc_auc': optimizar por AUC
# n_jobs=-1: usar todos los n√∫cleos de la CPU
grid_search_rf = GridSearchCV(estimator=model_rf_base, param_grid=param_grid_rf,
                              cv=5, scoring='roc_auc', n_jobs=-1, verbose=2)

# 4. Entrenar GridSearchCV para encontrar la mejor combinaci√≥n de par√°metros
# Esto puede tardar un tiempo considerable
grid_search_rf.fit(X_train, y_train)

# 5. Mostrar los mejores par√°metros encontrados y la mejor puntuaci√≥n de AUC
print(f"\nMejores par√°metros encontrados para Random Forest: {grid_search_rf.best_params_}")
print(f"Mejor AUC en validaci√≥n cruzada para Random Forest: {grid_search_rf.best_score_:.4f}")

# 6. Obtener el mejor modelo Random Forest
best_model_rf = grid_search_rf.best_estimator_

print("\n--- Evaluaci√≥n del Mejor Modelo Random Forest (GridSearchCV) ---")

# 7. Realizar predicciones con el mejor modelo sobre el conjunto de prueba
y_pred_best_rf = best_model_rf.predict(X_test)
y_pred_proba_best_rf = best_model_rf.predict_proba(X_test)[:, 1]

# 8. Calcular y mostrar m√©tricas de evaluaci√≥n para el mejor modelo
print(f"Precisi√≥n (Accuracy) del mejor Random Forest: {accuracy_score(y_test, y_pred_best_rf):.4f}")
print("\nReporte de Clasificaci√≥n del mejor Random Forest:")
print(classification_report(y_test, y_pred_best_rf))

auc_score_best_rf = roc_auc_score(y_test, y_pred_proba_best_rf)
print(f"AUC (Area Under the Curve) del mejor Random Forest: {auc_score_best_rf:.4f}")

# 9. Visualizar la Curva ROC del mejor modelo Random Forest
fpr_best_rf, tpr_best_rf, thresholds_best_rf = roc_curve(y_test, y_pred_proba_best_rf)

plt.figure(figsize=(8, 6))
plt.plot(fpr_best_rf, tpr_best_rf, color='darkgreen', lw=2, label=f'Curva ROC Mejor RF (AUC = {auc_score_best_rf:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', label='Clasificador Aleatorio')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Tasa de Falsos Positivos (FPR)')
plt.ylabel('Tasa de Verdaderos Positivos (TPR)')
plt.title('Curva ROC - Mejor Random Forest (GridSearchCV)')
plt.legend(loc="lower right")
plt.grid(True)
plt.show()

# Opcional: Visualizar la importancia de las caracter√≠sticas del mejor modelo Random Forest
importancia_caracteristicas_best_rf = pd.DataFrame({
    'Caracteristica': X.columns,
    'Importancia': best_model_rf.feature_importances_
}).sort_values(by='Importancia', ascending=False)

print("\nImportancia de las Caracter√≠sticas del Mejor Random Forest:")
print(importancia_caracteristicas_best_rf)

plt.figure(figsize=(10, 6))
sns.barplot(x='Importancia', y='Caracteristica', data=importancia_caracteristicas_best_rf)
plt.title('Importancia de las Caracter√≠sticas en el Mejor Random Forest')
plt.show()
```

## 10. Modelo 3: XGBoost (Extreme Gradient Boosting)

Finalmente, probaremos **XGBoost**, una de las implementaciones m√°s potentes y eficientes del algoritmo de *Gradient Boosting*. Es conocido por su alto rendimiento y es frecuentemente utilizado en competiciones de machine learning.

Para manejar el desbalance de clases, en lugar de `class_weight`, XGBoost utiliza el par√°metro `scale_pos_weight`, que se calcula como el ratio entre el n√∫mero de muestras de la clase negativa y la positiva.

```python
# Importar XGBoost
import xgboost as xgb
from sklearn.metrics import accuracy_score, classification_report, roc_auc_score, roc_curve
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd # Asegurarse de que pandas est√© importado

print("--- Entrenamiento del Modelo XGBoost ---")

# 1. Inicializar el modelo XGBoost
# Para clasificaci√≥n binaria, se usa 'binary:logistic' como objetivo.
# scale_pos_weight: importante para manejar el desbalance de clases.
#                   Se calcula como (n√∫mero de muestras negativas / n√∫mero de muestras positivas).
# random_state: para reproducibilidad
# use_label_encoder=False y eval_metric='logloss' son para evitar warnings en versiones recientes de XGBoost
neg_count = y_train.value_counts()[0]
pos_count = y_train.value_counts()[1]
scale_pos_weight_value = neg_count / pos_count

model_xgb = xgb.XGBClassifier(objective='binary:logistic',
                              eval_metric='logloss',
                              use_label_encoder=False,
                              random_state=42,
                              n_estimators=100, # N√∫mero de √°rboles
                              learning_rate=0.1, # Tasa de aprendizaje
                              max_depth=3, # Profundidad m√°xima de cada √°rbol
                              subsample=0.8, # Fracci√≥n de muestras para entrenar cada √°rbol
                              colsample_bytree=0.8, # Fracci√≥n de caracter√≠sticas para entrenar cada √°rbol
                              scale_pos_weight=scale_pos_weight_value # Manejo del desbalance de clases
                             )

# 2. Entrenar el modelo con los datos de entrenamiento
model_xgb.fit(X_train, y_train)

print("Modelo XGBoost entrenado exitosamente.")

print("\n--- Evaluaci√≥n del Modelo XGBoost ---")

# 3. Realizar predicciones sobre el conjunto de prueba
y_pred_xgb = model_xgb.predict(X_test)
y_pred_proba_xgb = model_xgb.predict_proba(X_test)[:, 1] # Probabilidades para la clase positiva (1)

# 4. Calcular y mostrar m√©tricas de evaluaci√≥n
print(f"Precisi√≥n (Accuracy) XGBoost: {accuracy_score(y_test, y_pred_xgb):.4f}")
print("\nReporte de Clasificaci√≥n XGBoost:")
print(classification_report(y_test, y_pred_xgb))

auc_score_xgb = roc_auc_score(y_test, y_pred_proba_xgb)
print(f"AUC (Area Under the Curve) XGBoost: {auc_score_xgb:.4f}")

# 5. Visualizar la Curva ROC del modelo XGBoost
fpr_xgb, tpr_xgb, thresholds_xgb = roc_curve(y_test, y_pred_proba_xgb)

plt.figure(figsize=(8, 6))
plt.plot(fpr_xgb, tpr_xgb, color='darkred', lw=2, label=f'Curva ROC XGBoost (AUC = {auc_score_xgb:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', label='Clasificador Aleatorio')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Tasa de Falsos Positivos (FPR)')
plt.ylabel('Tasa de Verdaderos Positivos (TPR)')
plt.title('Curva ROC - XGBoost')
plt.legend(loc="lower right")
plt.grid(True)
plt.show()

print("\n--- Importancia de las Caracter√≠sticas (XGBoost) ---")

# 6. Ver la importancia de las caracter√≠sticas
importancia_caracteristicas_xgb = pd.DataFrame({
    'Caracteristica': X.columns,
    'Importancia': model_xgb.feature_importances_
}).sort_values(by='Importancia', ascending=False)

print(importancia_caracteristicas_xgb)

plt.figure(figsize=(10, 6))
sns.barplot(x='Importancia', y='Caracteristica', data=importancia_caracteristicas_xgb)
plt.title('Importancia de las Caracter√≠sticas en XGBoost')
plt.show()
```

## 11. Optimizaci√≥n de XGBoost con GridSearchCV

El √∫ltimo paso de modelado es ajustar los hiperpar√°metros de XGBoost. Este proceso puede ser computacionalmente intensivo debido a la complejidad del algoritmo y la cantidad de par√°metros a probar (como `learning_rate`, `n_estimators`, `max_depth`, etc.). El objetivo es exprimir el m√°ximo rendimiento de este potente modelo.

```python
# Importar GridSearchCV para el ajuste de hiperpar√°metros
from sklearn.model_selection import GridSearchCV
import xgboost as xgb # Asegurarse de que est√© importado

print("--- Ajuste de Hiperpar√°metros para XGBoost (GridSearchCV) ---")

# 1. Definir los hiperpar√°metros que quieres probar y sus rangos para XGBoost
# Es importante ser cauteloso con el tama√±o de la cuadr√≠cula, ya que XGBoost puede ser computacionalmente intensivo.
param_grid_xgb = {
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.01, 0.05, 0.1],
    'max_depth': [3, 5, 7],
    'subsample': [0.7, 0.8, 0.9],
    'colsample_bytree': [0.7, 0.8, 0.9],
    'gamma': [0, 0.1, 0.2],
    # 'scale_pos_weight' se calcular√° din√°micamente, no lo incluimos en param_grid
}

# Calcular scale_pos_weight_value para el modelo base
neg_count = y_train.value_counts()[0]
pos_count = y_train.value_counts()[1]
scale_pos_weight_value = neg_count / pos_count

# 2. Inicializar el modelo XGBoost base con los par√°metros fijos (como random_state y manejo de desbalance)
model_xgb_base = xgb.XGBClassifier(objective='binary:logistic',
                                   eval_metric='logloss',
                                   use_label_encoder=False,
                                   random_state=42,
                                   scale_pos_weight=scale_pos_weight_value)

# 3. Configurar GridSearchCV para XGBoost
# cv=5: validaci√≥n cruzada de 5 pliegues
# scoring='roc_auc': optimizar por AUC
# n_jobs=-1: usar todos los n√∫cleos de la CPU
grid_search_xgb = GridSearchCV(estimator=model_xgb_base, param_grid=param_grid_xgb,
                               cv=5, scoring='roc_auc', n_jobs=-1, verbose=2)

# 4. Entrenar GridSearchCV para encontrar la mejor combinaci√≥n de par√°metros
# Esto puede tardar un tiempo considerable
grid_search_xgb.fit(X_train, y_train)

# 5. Mostrar los mejores par√°metros encontrados y la mejor puntuaci√≥n de AUC
print(f"\nMejores par√°metros encontrados para XGBoost: {grid_search_xgb.best_params_}")
print(f"Mejor AUC en validaci√≥n cruzada para XGBoost: {grid_search_xgb.best_score_:.4f}")

# 6. Obtener el mejor modelo XGBoost
best_model_xgb = grid_search_xgb.best_estimator_

print("\n--- Evaluaci√≥n del Mejor Modelo XGBoost (GridSearchCV) ---")

# 7. Realizar predicciones con el mejor modelo sobre el conjunto de prueba
y_pred_best_xgb = best_model_xgb.predict(X_test)
y_pred_proba_best_xgb = best_model_xgb.predict_proba(X_test)[:, 1]

# 8. Calcular y mostrar m√©tricas de evaluaci√≥n para el mejor modelo
print(f"Precisi√≥n (Accuracy) del mejor XGBoost: {accuracy_score(y_test, y_pred_best_xgb):.4f}")
print("\nReporte de Clasificaci√≥n del mejor XGBoost:")
print(classification_report(y_test, y_pred_best_xgb))

auc_score_best_xgb = roc_auc_score(y_test, y_pred_proba_best_xgb)
print(f"AUC (Area Under the Curve) del mejor XGBoost: {auc_score_best_xgb:.4f}")

# 9. Visualizar la Curva ROC del mejor modelo XGBoost
fpr_best_xgb, tpr_best_xgb, thresholds_best_xgb = roc_curve(y_test, y_pred_proba_best_xgb)

plt.figure(figsize=(8, 6))
plt.plot(fpr_best_xgb, tpr_best_xgb, color='darkred', lw=2, label=f'Curva ROC Mejor XGBoost (AUC = {auc_score_best_xgb:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', label='Clasificador Aleatorio')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Tasa de Falsos Positivos (FPR)')
plt.ylabel('Tasa de Verdaderos Positivos (TPR)')
plt.title('Curva ROC - Mejor XGBoost (GridSearchCV)')
plt.legend(loc="lower right")
plt.grid(True)
plt.show()

# Opcional: Visualizar la importancia de las caracter√≠sticas del mejor modelo XGBoost
importancia_caracteristicas_best_xgb = pd.DataFrame({
    'Caracteristica': X.columns,
    'Importancia': best_model_xgb.feature_importances_
}).sort_values(by='Importancia', ascending=False)

print("\nImportancia de las Caracter√≠sticas del Mejor XGBoost:")
print(importancia_caracteristicas_best_xgb)

plt.figure(figsize=(10, 6))
sns.barplot(x='Importancia', y='Caracteristica', data=importancia_caracteristicas_best_xgb)
plt.title('Importancia de las Caracter√≠sticas en el Mejor XGBoost')
plt.show()
```
