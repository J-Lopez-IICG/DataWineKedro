## 2. Limpieza y Verificación de Datos

Antes de cualquier análisis o modelado, es fundamental asegurar la calidad de los datos. En esta sección, realizaremos los siguientes pasos:
1.  **Verificar Valores Nulos:** Buscaremos y eliminaremos cualquier fila que contenga datos faltantes.
2.  **Verificar Duplicados:** Identificaremos y eliminaremos registros duplicados para evitar sesgos en el modelo.
3.  **Verificar Tipos de Datos:** Nos aseguraremos de que todas las columnas de características sean numéricas, lo cual es un requisito para los modelos que utilizaremos.

```python
# 1. Revisar si hay valores nulos en cada columna
print("Valores nulos por columna antes de la limpieza:")
print(data.isnull().sum())

# 2. Eliminar filas que contengan al menos un valor nulo
# .dropna() es el comando para eliminar filas (o columnas) con valores NaN
# 'inplace=True' modifica el DataFrame 'data' directamente
data.dropna(inplace=True)

print("\nValores nulos por columna después de la limpieza:")
print(data.isnull().sum())

# 3. Revisar si hay filas duplicadas
print(f"\nNúmero de filas duplicadas encontradas: {data.duplicated().sum()}")

# 4. Eliminar filas duplicadas
# 'inplace=True' modifica el DataFrame 'data' directamente
data.drop_duplicates(inplace=True)

print(f"Número de filas duplicadas después de la limpieza: {data.duplicated().sum()}")

# 5. Mostrar información general del DataFrame limpio
print("\nInformación del DataFrame después de la limpieza:")
data.info()

# 6. Verificar que todas las columnas sean numéricas (int o float)
print("\n--- Verificación de Tipos de Datos ---")
import pandas as pd
from pandas.api.types import is_numeric_dtype

non_numeric_columns = []
for column in data.columns:
    # Comprobamos si el tipo de dato de la columna NO es numérico
    if not is_numeric_dtype(data[column]):
        non_numeric_columns.append(column)

if not non_numeric_columns:
    print("¡Excelente! Todas las columnas tienen un formato numérico (int o float).")
else:
    print("¡Atención! Se encontraron columnas con formato no numérico:")
    for col in non_numeric_columns:
        print(f"- Columna: '{col}', Tipo: {data[col].dtype}")
```

## 3. Renombrar Columnas

Para mejorar la legibilidad y la coherencia del código, traduciremos los nombres de las columnas del inglés al español, utilizando un formato estándar de `snake_case` (por ejemplo, `fixed acidity` se convertirá en `acidez_fija`).

```python
# Crear un diccionario para mapear los nombres de las columnas de inglés a español
# Es una buena práctica usar minúsculas y guiones bajos para los nombres de las columnas
column_mapping = {
    'fixed acidity': 'Acidez_Fija',
    'volatile acidity': 'Acidez_Volatil',
    'citric acid': 'Acido_Citrico',
    'residual sugar': 'Azucar_Residual',
    'chlorides': 'Cloruros',
    'free sulfur dioxide': 'Dioxido_Azufre_Libre',
    'total sulfur dioxide': 'Dioxido_Azufre_Total',
    'density': 'Densidad',
    'pH': 'ph',
    'sulphates': 'Sulfatos',
    'alcohol': 'Alcohol',
    'quality': 'Calidad'  # No olvidar la variable de salida
}

# Renombrar las columnas del DataFrame
data.rename(columns=column_mapping, inplace=True)

# Verificar que los nombres de las columnas se han actualizado correctamente
print("Nombres de las columnas actualizados:")
print(data.columns)

# Opcional: Mostrar las primeras filas con los nuevos nombres
print("\nDataFrame con las nuevas columnas:")
data.head()
```

## 4. Ingeniería de Características: Creación de la Variable Objetivo

La columna `Calidad` original es una variable categórica ordinal con valores del 3 al 8. Para simplificar el problema y convertirlo en una tarea de clasificación binaria, crearemos una nueva columna llamada `Calidad_Binaria`:
-   **1 (Buena Calidad):** Si la calidad original es 7 o superior.
-   **0 (Mala Calidad):** Si la calidad original es inferior a 7.

Este enfoque es común cuando el objetivo es distinguir entre vinos "normales" y vinos "premium" o de alta calidad.

```python
# Crear la nueva columna 'Calidad_Binaria'
# Usaremos una condición booleana. Esto crea una serie de True/False.
# Al convertirla a entero con .astype(int), True se convierte en 1 y False en 0.
data['Calidad_Binaria'] = (data['Calidad'] >= 7).astype(int)

# Verificar la distribución de la nueva columna (opcional pero recomendado)
print("Distribución de la nueva columna 'Calidad_Binaria':")
print(data['Calidad_Binaria'].value_counts())

columnas_a_mantener = [
    'Acidez_Fija',
    'Acidez_Volatil',
    'Acido_Citrico',
    'Azucar_Residual',
    'Cloruros',
    'Dioxido_Azufre_Libre',
    'Dioxido_Azufre_Total',
    'Densidad',
    'ph',
    'Sulfatos',
    'Alcohol',
    #'Calidad',
    'Calidad_Binaria',
]
data_seleccionada = data[columnas_a_mantener].copy()

# Mostrar las primeras filas del nuevo DataFrame con las columnas seleccionadas
print("DataFrame con las columnas seleccionadas:")
print(data_seleccionada.head())

# Mostrar la información del nuevo DataFrame para confirmar las columnas y tipos de datos
print("\nInformación del DataFrame con columnas seleccionadas:")
data_seleccionada.info()
```

## 5. Preparación de Datos para el Modelado

En este paso, dividiremos nuestros datos para el entrenamiento y la evaluación del modelo:
1.  **Separación de Características (X) y Objetivo (y):** Aislaremos la columna `Calidad_Binaria` (nuestra variable objetivo `y`) del resto de las características (`X`).
2.  **División en Entrenamiento y Prueba:** Dividiremos el conjunto de datos en un 70% para entrenamiento (`X_train`, `y_train`) y un 30% para prueba (`X_test`, `y_test`). El conjunto de prueba no se utilizará durante el entrenamiento y servirá para evaluar el rendimiento final del modelo en datos no vistos. Usamos `stratify=y` para mantener la misma proporción de clases en ambos conjuntos, lo cual es crucial en datasets desbalanceados.

```python
# Importar las librerías necesarias para el modelado
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report, roc_auc_score, roc_curve
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd # Asegurarse de que pandas esté importado

print("--- Preparación de Datos para el Modelo ---")

# 1. Separar las características (X) de la variable objetivo (y)
# La variable objetivo es 'Calidad_Binaria'
X = data_seleccionada.drop('Calidad_Binaria', axis=1) # Todas las columnas excepto 'Calidad_Binaria'
y = data_seleccionada['Calidad_Binaria'] # Solo la columna 'Calidad_Binaria'

print(f"Dimensiones de X (características): {X.shape}")
print(f"Dimensiones de y (variable objetivo): {y.shape}")

# 2. Dividir los datos en conjuntos de entrenamiento y prueba
# test_size=0.30 significa que el 30% de los datos se usarán para prueba
# random_state=42 asegura que la división sea la misma cada vez que ejecutes el código
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42, stratify=y)

print(f"Dimensiones de X_train: {X_train.shape}")
print(f"Dimensiones de X_test: {X_test.shape}")
print(f"Dimensiones de y_train: {y_train.shape}")
print(f"Dimensiones de y_test: {y_test.shape}")
```

## 6. Modelo 1: Árbol de Decisión (Línea Base)

Comenzamos con un **Árbol de Decisión** como nuestro modelo de línea base. Es un modelo simple, rápido de entrenar y fácil de interpretar visualmente. Nos dará una primera idea del rendimiento que podemos esperar y de las características más importantes según este algoritmo.

```python
print("\n--- Entrenamiento del Modelo de Árbol de Decisión ---")

# 3. Inicializar el modelo de Árbol de Decisión
# random_state=42 para reproducibilidad
# Puedes ajustar otros parámetros como max_depth, min_samples_leaf, etc.
model_dt = DecisionTreeClassifier(random_state=42)

# 4. Entrenar el modelo con los datos de entrenamiento
model_dt.fit(X_train, y_train)

print("Modelo de Árbol de Decisión entrenado exitosamente.")

print("\n--- Evaluación del Modelo ---")

# 5. Realizar predicciones sobre el conjunto de prueba
y_pred = model_dt.predict(X_test)
y_pred_proba = model_dt.predict_proba(X_test)[:, 1] # type: ignore

# 6. Calcular y mostrar métricas de evaluación
print(f"Precisión (Accuracy): {accuracy_score(y_test, y_pred):.4f}")
print("\nReporte de Clasificación:")
print(classification_report(y_test, y_pred))

# Calcular y mostrar el AUC (Area Under the Curve)
auc_score = roc_auc_score(y_test, y_pred_proba)
print(f"AUC (Area Under the Curve): {auc_score:.4f}")

# 7. Visualizar la Curva ROC
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'Curva ROC (AUC = {auc_score:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', label='Clasificador Aleatorio')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Tasa de Falsos Positivos (FPR)')
plt.ylabel('Tasa de Verdaderos Positivos (TPR)')
plt.title('Curva ROC - Árbol de Decisión')
plt.legend(loc="lower right")
plt.grid(True)
plt.show()

print("\n--- Interpretación del Árbol de Decisión ---")
from sklearn.tree import plot_tree
plt.figure(figsize=(20,10))
plot_tree(model_dt, feature_names=X.columns, class_names=['Malo', 'Bueno'], filled=True, rounded=True)
plt.title("Árbol de Decisión")
plt.show()

# También puedes ver la importancia de las características
importancia_caracteristicas = pd.DataFrame({
    'Caracteristica': X.columns,
    'Importancia': model_dt.feature_importances_
}).sort_values(by='Importancia', ascending=False)

print("\nImportancia de las Características:")
print(importancia_caracteristicas)

plt.figure(figsize=(10, 6))
sns.barplot(x='Importancia', y='Caracteristica', data=importancia_caracteristicas)
plt.title('Importancia de las Características en el Árbol de Decisión')
plt.show()
```

## 7. Optimización del Árbol de Decisión con GridSearchCV

El rendimiento de un modelo depende en gran medida de sus **hiperparámetros**. Para mejorar nuestro Árbol de Decisión, utilizaremos `GridSearchCV`. Esta técnica prueba sistemáticamente una "cuadrícula" de diferentes combinaciones de hiperparámetros (como `max_depth` o `min_samples_leaf`) y, mediante validación cruzada, encuentra la combinación que maximiza el rendimiento, en nuestro caso, el `roc_auc`.

```python
# Importar GridSearchCV para el ajuste de hiperparámetros
from sklearn.model_selection import GridSearchCV

print("--- Ajuste de Hiperparámetros para el Árbol de Decisión (GridSearchCV) ---")

# 1. Definir los hiperparámetros que quieres probar y sus rangos
# Estos son algunos de los parámetros más comunes para un Árbol de Decisión
param_grid = {
    'max_depth': [3, 5, 7, 10, None], # Profundidad máxima del árbol. None significa sin límite.
    'min_samples_leaf': [1, 5, 10, 20], # Número mínimo de muestras requeridas para ser un nodo hoja.
    'min_samples_split': [2, 5, 10], # Número mínimo de muestras requeridas para dividir un nodo.
    'criterion': ['gini', 'entropy'] # Función para medir la calidad de una división.
}

# 2. Inicializar el modelo de Árbol de Decisión base
# Usamos el mismo random_state para reproducibilidad
model_dt_base = DecisionTreeClassifier(random_state=42)

# 3. Configurar GridSearchCV
# estimator: el modelo que queremos ajustar
# param_grid: el diccionario de parámetros a probar
# cv: número de pliegues para la validación cruzada (5 es un valor común)
# scoring: la métrica que queremos optimizar (roc_auc es ideal para clasificación binaria desbalanceada)
# n_jobs: -1 usa todos los núcleos de la CPU para acelerar el proceso
# verbose: 1 muestra el progreso
grid_search = GridSearchCV(estimator=model_dt_base, param_grid=param_grid,
                           cv=5, scoring='roc_auc', n_jobs=-1, verbose=1)

# 4. Entrenar GridSearchCV para encontrar la mejor combinación de parámetros
# Esto puede tardar un poco dependiendo del tamaño de tu param_grid y tus datos
grid_search.fit(X_train, y_train)

# 5. Mostrar los mejores parámetros encontrados y la mejor puntuación de AUC
print(f"\nMejores parámetros encontrados: {grid_search.best_params_}")
print(f"Mejor AUC en validación cruzada: {grid_search.best_score_:.4f}")

# 6. Obtener el mejor modelo (el Árbol de Decisión con los mejores parámetros)
best_model_dt = grid_search.best_estimator_

print("\n--- Evaluación del Mejor Modelo de Árbol de Decisión ---")

# 7. Realizar predicciones con el mejor modelo sobre el conjunto de prueba
y_pred_best = best_model_dt.predict(X_test)
y_pred_proba_best = best_model_dt.predict_proba(X_test)[:, 1] # type: ignore # Probabilidades para la clase positiva (1)

# 8. Calcular y mostrar métricas de evaluación para el mejor modelo
print(f"Precisión (Accuracy) del mejor modelo: {accuracy_score(y_test, y_pred_best):.4f}")
print("\nReporte de Clasificación del mejor modelo:")
print(classification_report(y_test, y_pred_best))

auc_score_best = roc_auc_score(y_test, y_pred_proba_best)
print(f"AUC (Area Under the Curve) del mejor modelo: {auc_score_best:.4f}")

# 9. Visualizar la Curva ROC del mejor modelo
fpr_best, tpr_best, thresholds_best = roc_curve(y_test, y_pred_proba_best)

plt.figure(figsize=(8, 6))
plt.plot(fpr_best, tpr_best, color='darkorange', lw=2, label=f'Curva ROC (AUC = {auc_score_best:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', label='Clasificador Aleatorio')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Tasa de Falsos Positivos (FPR)')
plt.ylabel('Tasa de Verdaderos Positivos (TPR)')
plt.title('Curva ROC - Mejor Árbol de Decisión (GridSearchCV)')
plt.legend(loc="lower right")
plt.grid(True)
plt.show()

# Opcional: Visualizar la importancia de las características del mejor modelo
importancia_caracteristicas_best = pd.DataFrame({
    'Caracteristica': X.columns,
    'Importancia': best_model_dt.feature_importances_
}).sort_values(by='Importancia', ascending=False)

print("\nImportancia de las Características del Mejor Modelo:")
print(importancia_caracteristicas_best)

plt.figure(figsize=(10, 6))
sns.barplot(x='Importancia', y='Caracteristica', data=importancia_caracteristicas_best)
plt.title('Importancia de las Características en el Mejor Árbol de Decisión')
plt.show()
```

## 8. Modelo 2: Random Forest

El siguiente modelo es el **Random Forest**, un método de *ensemble* que construye múltiples árboles de decisión durante el entrenamiento y combina sus resultados. Generalmente, esto conduce a un modelo más robusto, con mayor precisión y menor riesgo de sobreajuste que un único árbol de decisión.

Para manejar el desbalance de clases, utilizaremos el parámetro `class_weight='balanced'`, que ajusta los pesos de las muestras de forma inversamente proporcional a la frecuencia de sus clases.

```python
# Importar el clasificador Random Forest
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, roc_auc_score, roc_curve
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd # Asegurarse de que pandas esté importado

print("--- Entrenamiento del Modelo Random Forest ---")

# 1. Inicializar el modelo Random Forest
# n_estimators: número de árboles en el bosque (un valor común es 100 o más)
# random_state: para reproducibilidad
# class_weight='balanced': importante para manejar el desbalance de clases,
#                          ajusta automáticamente los pesos de las clases.
model_rf = RandomForestClassifier(n_estimators=100, random_state=42, class_weight='balanced')

# 2. Entrenar el modelo con los datos de entrenamiento
model_rf.fit(X_train, y_train)

print("Modelo Random Forest entrenado exitosamente.")

print("\n--- Evaluación del Modelo Random Forest ---")

# 3. Realizar predicciones sobre el conjunto de prueba
y_pred_rf = model_rf.predict(X_test)
y_pred_proba_rf = model_rf.predict_proba(X_test)[:, 1] # Probabilidades para la clase positiva (1)

# 4. Calcular y mostrar métricas de evaluación
print(f"Precisión (Accuracy) Random Forest: {accuracy_score(y_test, y_pred_rf):.4f}")
print("\nReporte de Clasificación Random Forest:")
print(classification_report(y_test, y_pred_rf))

auc_score_rf = roc_auc_score(y_test, y_pred_proba_rf)
print(f"AUC (Area Under the Curve) Random Forest: {auc_score_rf:.4f}")

# 5. Visualizar la Curva ROC del modelo Random Forest
fpr_rf, tpr_rf, thresholds_rf = roc_curve(y_test, y_pred_proba_rf)

plt.figure(figsize=(8, 6))
plt.plot(fpr_rf, tpr_rf, color='darkgreen', lw=2, label=f'Curva ROC Random Forest (AUC = {auc_score_rf:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', label='Clasificador Aleatorio')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Tasa de Falsos Positivos (FPR)')
plt.ylabel('Tasa de Verdaderos Positivos (TPR)')
plt.title('Curva ROC - Random Forest')
plt.legend(loc="lower right")
plt.grid(True)
plt.show()

print("\n--- Importancia de las Características (Random Forest) ---")

# 6. Ver la importancia de las características
importancia_caracteristicas_rf = pd.DataFrame({
    'Caracteristica': X.columns,
    'Importancia': model_rf.feature_importances_
}).sort_values(by='Importancia', ascending=False)

print(importancia_caracteristicas_rf)

plt.figure(figsize=(10, 6))
sns.barplot(x='Importancia', y='Caracteristica', data=importancia_caracteristicas_rf)
plt.title('Importancia de las Características en Random Forest')
plt.show()
```

## 9. Optimización de Random Forest con GridSearchCV

Al igual que con el Árbol de Decisión, ahora optimizaremos los hiperparámetros del Random Forest. La cuadrícula de búsqueda incluirá parámetros clave como el número de árboles (`n_estimators`), la profundidad máxima (`max_depth`) y los criterios de división. El objetivo es encontrar la configuración que nos dé el mejor rendimiento posible con este algoritmo.

```python
# Importar GridSearchCV para el ajuste de hiperparámetros
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier # Asegurarse de que esté importado

print("--- Ajuste de Hiperparámetros para Random Forest (GridSearchCV) ---")

# 1. Definir los hiperparámetros que quieres probar y sus rangos para Random Forest
# Es importante no hacer la cuadrícula demasiado grande, ya que Random Forest es más lento que un solo árbol.
param_grid_rf = {
    'n_estimators': [50, 100, 200], # Número de árboles
    'max_depth': [5, 10, 15, None], # Profundidad máxima de cada árbol
    'min_samples_leaf': [1, 5, 10], # Mínimo de muestras en un nodo hoja
    'min_samples_split': [2, 5, 10], # Mínimo de muestras para dividir un nodo
    'criterion': ['gini', 'entropy'], # Criterio de división
    'class_weight': ['balanced'] # Mantener el manejo del desbalance de clases
}

# 2. Inicializar el modelo Random Forest base
# Usamos el mismo random_state para reproducibilidad
model_rf_base = RandomForestClassifier(random_state=42)

# 3. Configurar GridSearchCV para Random Forest
# cv=5: validación cruzada de 5 pliegues
# scoring='roc_auc': optimizar por AUC
# n_jobs=-1: usar todos los núcleos de la CPU
grid_search_rf = GridSearchCV(estimator=model_rf_base, param_grid=param_grid_rf,
                              cv=5, scoring='roc_auc', n_jobs=-1, verbose=2)

# 4. Entrenar GridSearchCV para encontrar la mejor combinación de parámetros
# Esto puede tardar un tiempo considerable
grid_search_rf.fit(X_train, y_train)

# 5. Mostrar los mejores parámetros encontrados y la mejor puntuación de AUC
print(f"\nMejores parámetros encontrados para Random Forest: {grid_search_rf.best_params_}")
print(f"Mejor AUC en validación cruzada para Random Forest: {grid_search_rf.best_score_:.4f}")

# 6. Obtener el mejor modelo Random Forest
best_model_rf = grid_search_rf.best_estimator_

print("\n--- Evaluación del Mejor Modelo Random Forest (GridSearchCV) ---")

# 7. Realizar predicciones con el mejor modelo sobre el conjunto de prueba
y_pred_best_rf = best_model_rf.predict(X_test)
y_pred_proba_best_rf = best_model_rf.predict_proba(X_test)[:, 1]

# 8. Calcular y mostrar métricas de evaluación para el mejor modelo
print(f"Precisión (Accuracy) del mejor Random Forest: {accuracy_score(y_test, y_pred_best_rf):.4f}")
print("\nReporte de Clasificación del mejor Random Forest:")
print(classification_report(y_test, y_pred_best_rf))

auc_score_best_rf = roc_auc_score(y_test, y_pred_proba_best_rf)
print(f"AUC (Area Under the Curve) del mejor Random Forest: {auc_score_best_rf:.4f}")

# 9. Visualizar la Curva ROC del mejor modelo Random Forest
fpr_best_rf, tpr_best_rf, thresholds_best_rf = roc_curve(y_test, y_pred_proba_best_rf)

plt.figure(figsize=(8, 6))
plt.plot(fpr_best_rf, tpr_best_rf, color='darkgreen', lw=2, label=f'Curva ROC Mejor RF (AUC = {auc_score_best_rf:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', label='Clasificador Aleatorio')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Tasa de Falsos Positivos (FPR)')
plt.ylabel('Tasa de Verdaderos Positivos (TPR)')
plt.title('Curva ROC - Mejor Random Forest (GridSearchCV)')
plt.legend(loc="lower right")
plt.grid(True)
plt.show()

# Opcional: Visualizar la importancia de las características del mejor modelo Random Forest
importancia_caracteristicas_best_rf = pd.DataFrame({
    'Caracteristica': X.columns,
    'Importancia': best_model_rf.feature_importances_
}).sort_values(by='Importancia', ascending=False)

print("\nImportancia de las Características del Mejor Random Forest:")
print(importancia_caracteristicas_best_rf)

plt.figure(figsize=(10, 6))
sns.barplot(x='Importancia', y='Caracteristica', data=importancia_caracteristicas_best_rf)
plt.title('Importancia de las Características en el Mejor Random Forest')
plt.show()
```

## 10. Modelo 3: XGBoost (Extreme Gradient Boosting)

Finalmente, probaremos **XGBoost**, una de las implementaciones más potentes y eficientes del algoritmo de *Gradient Boosting*. Es conocido por su alto rendimiento y es frecuentemente utilizado en competiciones de machine learning.

Para manejar el desbalance de clases, en lugar de `class_weight`, XGBoost utiliza el parámetro `scale_pos_weight`, que se calcula como el ratio entre el número de muestras de la clase negativa y la positiva.

```python
# Importar XGBoost
import xgboost as xgb
from sklearn.metrics import accuracy_score, classification_report, roc_auc_score, roc_curve
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd # Asegurarse de que pandas esté importado

print("--- Entrenamiento del Modelo XGBoost ---")

# 1. Inicializar el modelo XGBoost
# Para clasificación binaria, se usa 'binary:logistic' como objetivo.
# scale_pos_weight: importante para manejar el desbalance de clases.
#                   Se calcula como (número de muestras negativas / número de muestras positivas).
# random_state: para reproducibilidad
# use_label_encoder=False y eval_metric='logloss' son para evitar warnings en versiones recientes de XGBoost
neg_count = y_train.value_counts()[0]
pos_count = y_train.value_counts()[1]
scale_pos_weight_value = neg_count / pos_count

model_xgb = xgb.XGBClassifier(objective='binary:logistic',
                              eval_metric='logloss',
                              use_label_encoder=False,
                              random_state=42,
                              n_estimators=100, # Número de árboles
                              learning_rate=0.1, # Tasa de aprendizaje
                              max_depth=3, # Profundidad máxima de cada árbol
                              subsample=0.8, # Fracción de muestras para entrenar cada árbol
                              colsample_bytree=0.8, # Fracción de características para entrenar cada árbol
                              scale_pos_weight=scale_pos_weight_value # Manejo del desbalance de clases
                             )

# 2. Entrenar el modelo con los datos de entrenamiento
model_xgb.fit(X_train, y_train)

print("Modelo XGBoost entrenado exitosamente.")

print("\n--- Evaluación del Modelo XGBoost ---")

# 3. Realizar predicciones sobre el conjunto de prueba
y_pred_xgb = model_xgb.predict(X_test)
y_pred_proba_xgb = model_xgb.predict_proba(X_test)[:, 1] # Probabilidades para la clase positiva (1)

# 4. Calcular y mostrar métricas de evaluación
print(f"Precisión (Accuracy) XGBoost: {accuracy_score(y_test, y_pred_xgb):.4f}")
print("\nReporte de Clasificación XGBoost:")
print(classification_report(y_test, y_pred_xgb))

auc_score_xgb = roc_auc_score(y_test, y_pred_proba_xgb)
print(f"AUC (Area Under the Curve) XGBoost: {auc_score_xgb:.4f}")

# 5. Visualizar la Curva ROC del modelo XGBoost
fpr_xgb, tpr_xgb, thresholds_xgb = roc_curve(y_test, y_pred_proba_xgb)

plt.figure(figsize=(8, 6))
plt.plot(fpr_xgb, tpr_xgb, color='darkred', lw=2, label=f'Curva ROC XGBoost (AUC = {auc_score_xgb:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', label='Clasificador Aleatorio')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Tasa de Falsos Positivos (FPR)')
plt.ylabel('Tasa de Verdaderos Positivos (TPR)')
plt.title('Curva ROC - XGBoost')
plt.legend(loc="lower right")
plt.grid(True)
plt.show()

print("\n--- Importancia de las Características (XGBoost) ---")

# 6. Ver la importancia de las características
importancia_caracteristicas_xgb = pd.DataFrame({
    'Caracteristica': X.columns,
    'Importancia': model_xgb.feature_importances_
}).sort_values(by='Importancia', ascending=False)

print(importancia_caracteristicas_xgb)

plt.figure(figsize=(10, 6))
sns.barplot(x='Importancia', y='Caracteristica', data=importancia_caracteristicas_xgb)
plt.title('Importancia de las Características en XGBoost')
plt.show()
```

## 11. Optimización de XGBoost con GridSearchCV

El último paso de modelado es ajustar los hiperparámetros de XGBoost. Este proceso puede ser computacionalmente intensivo debido a la complejidad del algoritmo y la cantidad de parámetros a probar (como `learning_rate`, `n_estimators`, `max_depth`, etc.). El objetivo es exprimir el máximo rendimiento de este potente modelo.

```python
# Importar GridSearchCV para el ajuste de hiperparámetros
from sklearn.model_selection import GridSearchCV
import xgboost as xgb # Asegurarse de que esté importado

print("--- Ajuste de Hiperparámetros para XGBoost (GridSearchCV) ---")

# 1. Definir los hiperparámetros que quieres probar y sus rangos para XGBoost
# Es importante ser cauteloso con el tamaño de la cuadrícula, ya que XGBoost puede ser computacionalmente intensivo.
param_grid_xgb = {
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.01, 0.05, 0.1],
    'max_depth': [3, 5, 7],
    'subsample': [0.7, 0.8, 0.9],
    'colsample_bytree': [0.7, 0.8, 0.9],
    'gamma': [0, 0.1, 0.2],
    # 'scale_pos_weight' se calculará dinámicamente, no lo incluimos en param_grid
}

# Calcular scale_pos_weight_value para el modelo base
neg_count = y_train.value_counts()[0]
pos_count = y_train.value_counts()[1]
scale_pos_weight_value = neg_count / pos_count

# 2. Inicializar el modelo XGBoost base con los parámetros fijos (como random_state y manejo de desbalance)
model_xgb_base = xgb.XGBClassifier(objective='binary:logistic',
                                   eval_metric='logloss',
                                   use_label_encoder=False,
                                   random_state=42,
                                   scale_pos_weight=scale_pos_weight_value)

# 3. Configurar GridSearchCV para XGBoost
# cv=5: validación cruzada de 5 pliegues
# scoring='roc_auc': optimizar por AUC
# n_jobs=-1: usar todos los núcleos de la CPU
grid_search_xgb = GridSearchCV(estimator=model_xgb_base, param_grid=param_grid_xgb,
                               cv=5, scoring='roc_auc', n_jobs=-1, verbose=2)

# 4. Entrenar GridSearchCV para encontrar la mejor combinación de parámetros
# Esto puede tardar un tiempo considerable
grid_search_xgb.fit(X_train, y_train)

# 5. Mostrar los mejores parámetros encontrados y la mejor puntuación de AUC
print(f"\nMejores parámetros encontrados para XGBoost: {grid_search_xgb.best_params_}")
print(f"Mejor AUC en validación cruzada para XGBoost: {grid_search_xgb.best_score_:.4f}")

# 6. Obtener el mejor modelo XGBoost
best_model_xgb = grid_search_xgb.best_estimator_

print("\n--- Evaluación del Mejor Modelo XGBoost (GridSearchCV) ---")

# 7. Realizar predicciones con el mejor modelo sobre el conjunto de prueba
y_pred_best_xgb = best_model_xgb.predict(X_test)
y_pred_proba_best_xgb = best_model_xgb.predict_proba(X_test)[:, 1]

# 8. Calcular y mostrar métricas de evaluación para el mejor modelo
print(f"Precisión (Accuracy) del mejor XGBoost: {accuracy_score(y_test, y_pred_best_xgb):.4f}")
print("\nReporte de Clasificación del mejor XGBoost:")
print(classification_report(y_test, y_pred_best_xgb))

auc_score_best_xgb = roc_auc_score(y_test, y_pred_proba_best_xgb)
print(f"AUC (Area Under the Curve) del mejor XGBoost: {auc_score_best_xgb:.4f}")

# 9. Visualizar la Curva ROC del mejor modelo XGBoost
fpr_best_xgb, tpr_best_xgb, thresholds_best_xgb = roc_curve(y_test, y_pred_proba_best_xgb)

plt.figure(figsize=(8, 6))
plt.plot(fpr_best_xgb, tpr_best_xgb, color='darkred', lw=2, label=f'Curva ROC Mejor XGBoost (AUC = {auc_score_best_xgb:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', label='Clasificador Aleatorio')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Tasa de Falsos Positivos (FPR)')
plt.ylabel('Tasa de Verdaderos Positivos (TPR)')
plt.title('Curva ROC - Mejor XGBoost (GridSearchCV)')
plt.legend(loc="lower right")
plt.grid(True)
plt.show()

# Opcional: Visualizar la importancia de las características del mejor modelo XGBoost
importancia_caracteristicas_best_xgb = pd.DataFrame({
    'Caracteristica': X.columns,
    'Importancia': best_model_xgb.feature_importances_
}).sort_values(by='Importancia', ascending=False)

print("\nImportancia de las Características del Mejor XGBoost:")
print(importancia_caracteristicas_best_xgb)

plt.figure(figsize=(10, 6))
sns.barplot(x='Importancia', y='Caracteristica', data=importancia_caracteristicas_best_xgb)
plt.title('Importancia de las Características en el Mejor XGBoost')
plt.show()
```
